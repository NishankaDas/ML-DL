Image Negative Transformation

An **image negative transformation** inverts the intensity values of an image, making **dark regions appear light** and **light regions appear dark**. This transformation is commonly used in image processing to enhance details in images where dark areas contain important information.

The transformation is defined by the following equation:

\[
s = L - 1 - r
\]

where:

- \( s \) = output (transformed) pixel intensity
- \( r \) = original pixel intensity
- \( L \) = total number of intensity levels

For an **8-bit grayscale image**, the number of intensity levels is:


This transformation effectively reverses the grayscale intensity values, producing the negative of the original image.

````python
img_neg = 255 - img
````

````python
plt.imshow(img_neg)
plt.axis('off')
plt.show()
````

 Log Transformation

A **log transformation** expands dark pixel values while compressing bright pixel values, making details in darker regions more visible.

The transformation is given by:
$$
s = c\log(1 + r)
$$
where:

- \( s \) = output pixel value
- \( r \) = input pixel value
- \( c \) = scaling constant
```

````python
c = 255/ np.log(1 + np.max(img))
log_img = c*(np.log(img +1))
log_img = np.array(log_img, dtype = np.uint8)
````

````python
plt.imshow(log_img)
plt.axis('off')
plt.show()
````

 Exponential Transformation

An **exponential transformation** enhances brighter regions while compressing darker regions.

$$
s = c\left[ \left( 1+ \alpha \right)^r - 1\right]
$$

where:

- \( s \) = output pixel value
- \( r \) = input pixel value (typically normalized)
- \( c \) = scaling constant

````python
alpha = 0.8  # Change this value as needed
r = 0.4
img_norm = img / 255.0
c = 255 / ((1 + alpha)**r - 1)
exp_img = c * ((1 + alpha)**img_norm - 1)
exp_img = np.array(exp_img, dtype=np.uint8)
````

````python
plt.imshow(exp_img)
plt.axis('off')
plt.show()
````

Power Transformation 

An **power transformation** enhances brighter regions while compressing darker regions.

$$
s = cr^\gamma
$$

where:

- \( s \) = output pixel value
- \( r \) = input pixel value (typically normalized)
- \( c \) = scaling constant

The illustration of the image intensity for just varring the $\gamma$ is

````python
gamma = 8
img_norm = img / 255.0
c = 255
power_img = c * (img_norm ** gamma)
power_img = np.array(power_img, dtype=np.uint8)
````

````python
plt.imshow(power_img)
plt.axis('off')
plt.show()
````
