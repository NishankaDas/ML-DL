```python
import numpy as np
import matplotlib.pyplot as plt

# Values of p from 0 to 1 with step 0.1
p = np.arange(0, 1.01, 0.01)

# Compute binary entropy
H = np.zeros_like(p)

for i in range(len(p)):
    if p[i] == 0 or p[i] == 1:
        H[i] = 0  # Define entropy at p=0 and p=1
    else:
        H[i] = -(p[i] * np.log2(p[i]) +
                 (1 - p[i]) * np.log2(1 - p[i]))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(p, H, marker='o', linewidth=2, color='blue')
plt.title("Binary Entropy Function")
plt.xlabel("Probability (p)")
plt.ylabel("H(p)")
plt.grid(True)
plt.xticks(np.arange(0, 1.1, 0.1))
plt.ylim(0, 1.1)
plt.show()
```
