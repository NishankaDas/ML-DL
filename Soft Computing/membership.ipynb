# Descriptive implementation of diffrent Membership Functions

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

```python
values = [random.randint(20,230) for _ in range(100)]
plotter = [i for i in range(0,100)]
```

Tringular Membership function

For any given input x, the membership degeree $\mu_{A}(x)$ is calculated using following piecwise linear function
$$
\mu_{A}(x)=
\begin{cases}
0, & \text{if } x \le a \\
\frac{x-a}{b-a}, & a < x \le b \\
\frac{c-x}{c-b}, &b < x \le c \\
0, & \text{if } x > c \\
\end{cases}
$$

```python
def tmf(x,a,b,c):
    if x<= a:
        return 0
    if x> a and x <= b:
        return ((x-a)/(b-a))
    if b > x and x <= c:
        return ((c-x)/(c-b))
    else:
        return 0
```

```python
hist = []
a = 30
b = 120
c = 200
for i in values:
    hist.append(tmf(i,a,b,c))


normalized = []
for i in values:
    normalized.append(i/230)
plt.scatter(normalized, plotter, color = 'red', label = 'Normalized Value of Random Values')
plt.scatter(hist, plotter, color = 'blue', label = 'Traingular Membership Function')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Traingular Membership Function")
plt.legend()
plt.grid(True)

plt.show()
```

Trapizoidal Membership Function 

For any given input \(x\), the membership degree \(\mu_A(x)\) is calculated using the following piecewise linear function:

$$
\mu_A(x)=
\begin{cases}
0, & \text{if } x \le a \\[6pt]
\dfrac{x-a}{b-a}, & a < x \le b \\[10pt]
1, & b < x \le c \\[6pt]
\dfrac{d-x}{d-c}, & c < x \le d \\[10pt]
0, & \text{if } x > d
\end{cases}
$$

where

$$
a < b \le c < d
$$

Here,

- \(a\): Left foot of the trapezoid
- \(b\): Left shoulder
- \(c\): Right shoulder
- \(d\): Right foot


```python
def trapmf(x, a, b, c, d):
    if x <= a:
        return 0

    elif a < x <= b:
        return (x - a) / (b - a)

    elif b < x <= c:
        return 1

    elif c < x <= d:
        return (d - x) / (d - c)

    else:
        return 0
```

```python
hist = []

a = 30
b = 80
c = 150
d = 200

for i in values:
    hist.append(trapmf(i, a, b, c, d))

normalized = []
for i in values:
    normalized.append(i / 230)

plt.scatter(normalized, plotter,
            color='red',
            label='Normalized Values')

plt.scatter(hist, plotter,
            color='blue',
            label='Trapezoidal Membership Function')

plt.xlabel("Membership Degree")
plt.ylabel("Sample Index")
plt.title("Trapezoidal Membership Function")
plt.legend()
plt.grid(True)
plt.show()
```

 Gaussian Membership Function

For any given input \(x\), the membership degree \(\mu_A(x)\) is calculated using the Gaussian function:

$$
\mu_A(x)=\exp\left(-\frac{(x-c)^2}{2\sigma^2}\right)
$$

where

- \(c\): Center (mean) of the Gaussian curve.
- \( $\sigma$ \): Standard deviation, which controls the width (spread) of the curve.

Here,

- The membership degree is maximum, i.e.,\($\mu_A(c)=1$\).
- As the distance between \(x\) and \(c\) increases, the membership degree decreases smoothly toward 0.
- A larger value of \( $\sigma$ \) produces a wider Gaussian curve, while a smaller value produces a narrower curve.

```python
def gaussmf(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))
```

```python
hist = []

c = 120
sigma = 30

for i in values:
    hist.append(gaussmf(i, c, sigma))

normalized = []
for i in values:
    normalized.append(i / 230)

plt.scatter(normalized, plotter,
            color='red',
            label='Normalized Values')

plt.scatter(hist, plotter,
            color='blue',
            label='Gaussian Membership Function')

plt.xlabel("Membership Degree")
plt.ylabel("Sample Index")
plt.title("Gaussian Membership Function")
plt.legend()
plt.grid(True)
plt.show()
```

Generalized Bell Membership Function

For any given input \(x\), the membership degree \(\mu_A(x)\) is calculated using the generalized bell function:

$$
\mu_A(x)=\frac{1}{1+\left|\dfrac{x-c}{a}\right|^{2b}}
$$

where

- \(a\): Controls the width of the bell curve.
- \(b\): Controls the slope (steepness) of the curve.
- \(c\): Center of the bell curve.

Here,

- The maximum membership occurs at \(x=c\), where

$$
\mu_A(c)=1.
$$

- Increasing \(a\) widens the bell curve.
- Increasing \(b\) makes the curve steeper around the center.

```python
def gbellmf(x, a, b, c):
    return 1 / (1 + abs((x - c) / a) ** (2 * b))
```

```python
hist = []

# Parameters
a = 40      # Width
b = 2       # Slope
c = 120     # Center

# Compute membership values
for i in values:
    hist.append(gbellmf(i, a, b, c))

# Normalize original values (optional)
normalized = []
for i in values:
    normalized.append(i / 230)

# Plot
plt.scatter(normalized, plotter,
            color='red',
            label='Normalized Values')

plt.scatter(hist, plotter,
            color='blue',
            label='Generalized Bell Membership Function')

plt.xlabel("Membership Degree")
plt.ylabel("Sample Index")
plt.title("Generalized Bell Membership Function")
plt.legend()
plt.grid(True)
plt.show()
```

Sigmoid Membership Function

For any given input \(x\), the membership degree \(\mu_A(x)\) is calculated using the sigmoid function:

$$
\mu_A(x)=\frac{1}{1+e^{-a(x-c)}}
$$

where

- \(a\): Controls the slope (steepness) of the sigmoid curve.
- \(c\): Center (crossover point) of the sigmoid curve.

Here,

- The membership degree is

$$
\mu_A(c)=0.5.
$$

- If \(a>0\), the function is monotonically increasing.
- If \(a<0\), the function is monotonically decreasing.
- A larger value of \(|a|\) produces a steeper transition.

```python
def sigmf(x, a, c):
    return 1 / (1 + np.exp(-a * (x - c)))
```

```python
hist = []

# Parameters
a = 0.08      # Slope
c = 120       # Center

# Compute membership values
for i in values:
    hist.append(sigmf(i, a, c))

# Normalize original values (optional)
normalized = []
for i in values:
    normalized.append(i / 230)

# Plot
plt.scatter(normalized, plotter,
            color='red',
            label='Normalized Values')

plt.scatter(hist, plotter,
            color='blue',
            label='Sigmoid Membership Function')

plt.xlabel("Membership Degree")
plt.ylabel("Sample Index")
plt.title("Sigmoid Membership Function")
plt.legend()
plt.grid(True)
plt.show()
```

Z-Shaped Membership Function

For any given input \(x\), the membership degree \(\mu_A(x)\) is calculated using the following piecewise function:

$$
\mu_A(x)=
\begin{cases}
1, & \text{if } x \le a \\[8pt]
1-2\left(\dfrac{x-a}{b-a}\right)^2, & a < x \le \dfrac{a+b}{2} \\[12pt]
2\left(\dfrac{x-b}{b-a}\right)^2, & \dfrac{a+b}{2} < x \le b \\[12pt]
0, & \text{if } x>b
\end{cases}
$$

where

- \(a\): Left endpoint.
- \(b\): Right endpoint.

Here,

- The membership degree is maximum (\($\mu_A(x)=1$\)) for \($x\le a$\).
- The membership degree decreases smoothly from 1 to 0 over the interval \([a,b]\).
- The membership degree becomes zero for \(x>b\).

```python
def zmf(x, a, b):

    if x <= a:
        return 1

    elif a < x <= (a + b) / 2:
        return 1 - 2 * ((x - a) / (b - a))**2

    elif (a + b) / 2 < x <= b:
        return 2 * ((x - b) / (b - a))**2

    else:
        return 0
```

```python
hist = []

# Parameters
a = 50
b = 180

# Compute membership values
for i in values:
    hist.append(zmf(i, a, b))

# Normalize original values (optional)
normalized = []
for i in values:
    normalized.append(i / 230)

# Plot
plt.scatter(normalized, plotter,
            color='red',
            label='Normalized Values')

plt.scatter(hist, plotter,
            color='blue',
            label='Z-Shaped Membership Function')

plt.xlabel("Membership Degree")
plt.ylabel("Sample Index")
plt.title("Z-Shaped Membership Function")
plt.legend()
plt.grid(True)
plt.show()
```

S-Shaped Membership Function

For any given input \(x\), the membership degree \(\mu_A(x)\) is calculated using the following piecewise function:

$$
\mu_A(x)=
\begin{cases}
0, & \text{if } x \le a \\[8pt]
2\left(\dfrac{x-a}{b-a}\right)^2, & a < x \le \dfrac{a+b}{2} \\[12pt]
1-2\left(\dfrac{x-b}{b-a}\right)^2, & \dfrac{a+b}{2} < x \le b \\[12pt]
1, & \text{if } x>b
\end{cases}
$$

where

- \(a\): Left endpoint.
- \(b\): Right endpoint.

Here,

- The membership degree is zero for \($x\le a$\).
- The membership degree increases smoothly from 0 to 1 over the interval \([a,b]\).
- The membership degree becomes one for \(x>b\).

```python
def smf(x, a, b):

    if x <= a:
        return 0

    elif a < x <= (a + b) / 2:
        return 2 * ((x - a) / (b - a))**2

    elif (a + b) / 2 < x <= b:
        return 1 - 2 * ((x - b) / (b - a))**2

    else:
        return 1
```

```python
hist = []

# Parameters
a = 50
b = 180

# Compute membership values
for i in values:
    hist.append(smf(i, a, b))

# Normalize original values (optional)
normalized = []
for i in values:
    normalized.append(i / 230)

# Plot
plt.scatter(normalized, plotter,
            color='red',
            label='Normalized Values')

plt.scatter(hist, plotter,
            color='blue',
            label='S-Shaped Membership Function')

plt.xlabel("Membership Degree")
plt.ylabel("Sample Index")
plt.title("S-Shaped Membership Function")
plt.legend()
plt.grid(True)
plt.show()
```

Pi-Shaped Membership Function

For any given input \(x\), the membership degree \(\mu_A(x)\) is calculated using the following piecewise function:

$$
\mu_A(x)=
\begin{cases}
0, & \text{if } x \le a \\[8pt]

2\left(\dfrac{x-a}{b-a}\right)^2,
& a < x \le \dfrac{a+b}{2} \\[12pt]

1-2\left(\dfrac{x-b}{b-a}\right)^2,
& \dfrac{a+b}{2} < x \le b \\[12pt]

1,
& b < x \le c \\[8pt]

1-2\left(\dfrac{x-c}{d-c}\right)^2,
& c < x \le \dfrac{c+d}{2} \\[12pt]

2\left(\dfrac{x-d}{d-c}\right)^2,
& \dfrac{c+d}{2} < x \le d \\[12pt]

0,
& \text{if } x>d
\end{cases}
$$

where

- \(a\): Left foot
- \(b\): Left shoulder
- \(c\): Right shoulder
- \(d\): Right foot

Here,

- The membership degree increases smoothly from 0 to 1 over \([a,b]\).
- It remains equal to 1 over \([b,c]\).
- It decreases smoothly from 1 to 0 over \([c,d]\).

```python
def pimf(x, a, b, c, d):

    if x <= a:
        return 0

    elif a < x <= (a + b) / 2:
        return 2 * ((x - a) / (b - a))**2

    elif (a + b) / 2 < x <= b:
        return 1 - 2 * ((x - b) / (b - a))**2

    elif b < x <= c:
        return 1

    elif c < x <= (c + d) / 2:
        return 1 - 2 * ((x - c) / (d - c))**2

    elif (c + d) / 2 < x <= d:
        return 2 * ((x - d) / (d - c))**2

    else:
        return 0
```

```python
hist = []

# Parameters
a = 30
b = 80
c = 150
d = 200

# Compute membership values
for i in values:
    hist.append(pimf(i, a, b, c, d))

# Normalize original values (optional)
normalized = []
for i in values:
    normalized.append(i / 230)

# Plot
plt.scatter(normalized, plotter,
            color='red',
            label='Normalized Values')

plt.scatter(hist, plotter,
            color='blue',
            label='Pi-Shaped Membership Function')

plt.xlabel("Membership Degree")
plt.ylabel("Sample Index")
plt.title("Pi-Shaped Membership Function")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0,100,1000)

fig,axs = plt.subplots(4,2,figsize=(14,16))

# -------------------------------------------------
# 1. Triangular
# -------------------------------------------------

ax = axs[0,0]

params = [(0,0,25),
          (0,25,50),
          (25,50,75),
          (50,75,100),
          (75,100,100)]

labels=["VL","L","M","H","VH"]

for p,l in zip(params,labels):
    ax.plot(x,[tmf(i,*p) for i in x],label=l)

ax.set_title("Triangular Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------
# 2. Trapezoidal
# -------------------------------------------------

ax=axs[0,1]

params=[(0,0,15,30),
        (15,30,40,55),
        (40,50,60,70),
        (55,70,85,100),
        (70,85,100,100)]

for p,l in zip(params,labels):
    ax.plot(x,[trapmf(i,*p) for i in x],label=l)

ax.set_title("Trapezoidal Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------
# 3. Gaussian
# -------------------------------------------------

ax=axs[1,0]

centers=[10,30,50,70,90]

for c,l in zip(centers,labels):
    ax.plot(x,[gaussmf(i,c,12) for i in x],label=l)

ax.set_title("Gaussian Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------
# 4. Generalized Bell
# -------------------------------------------------

ax=axs[1,1]

for c,l in zip(centers,labels):
    ax.plot(x,[gbellmf(i,12,2,c) for i in x],label=l)

ax.set_title("Generalized Bell Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------
# 5. Sigmoid
# -------------------------------------------------

ax=axs[2,0]

params=[(-0.25,20),
        (-0.15,40),
        (0.15,50),
        (0.15,60),
        (0.25,80)]

for p,l in zip(params,labels):
    ax.plot(x,[sigmf(i,p[0],p[1]) for i in x],label=l)

ax.set_title("Sigmoid Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------
# 6. Z Shape
# -------------------------------------------------

ax=axs[2,1]

params=[(10,30),
        (25,45),
        (40,60),
        (55,75),
        (70,90)]

for p,l in zip(params,labels):
    ax.plot(x,[zmf(i,*p) for i in x],label=l)

ax.set_title("Z-Shaped Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------
# 7. S Shape
# -------------------------------------------------

ax=axs[3,0]

params=[(0,20),
        (15,35),
        (40,60),
        (65,85),
        (80,100)]

for p,l in zip(params,labels):
    ax.plot(x,[smf(i,*p) for i in x],label=l)

ax.set_title("S-Shaped Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------
# 8. Pi Shape
# -------------------------------------------------

ax=axs[3,1]

params=[(0,10,20,30),
        (15,25,35,45),
        (35,45,55,65),
        (55,65,75,85),
        (70,80,90,100)]

for p,l in zip(params,labels):
    ax.plot(x,[pimf(i,*p) for i in x],label=l)

ax.set_title("Pi-Shaped Membership Function")
ax.grid()
ax.legend()

# -------------------------------------------------

for ax in axs.ravel():
    ax.set_xlim(0,100)
    ax.set_ylim(-0.05,1.05)
    ax.set_xlabel("Universe of Discourse")
    ax.set_ylabel(r"$\mu(x)$")

plt.tight_layout()
plt.show()
```
