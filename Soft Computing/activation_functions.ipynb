Importing libraries 

````python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
````

Generating Random values for the plot

````python
v = []
for i in range(0,25):
    v.append(round(random.uniform(-1,1),4))
numbers = list(range(1, 26))
````

Binary Step Function

The mathematical representation of the Binary Step Function is:

$$
f(x)=
\begin{cases}
0, & \text{if } x < 0 \\
1, & \text{if } x \ge 0
\end{cases}
$$

````python
def bsf(u):
    result = []
    for x in u:
        if x < 0:
            result.append(0)
        else:
            result.append(1)
    return result
````

````python
g = bsf(v)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Binary Step Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Binary Step Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

 Sigmoid Activation Function

It is mathematically defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Where:

- $\sigma(x)$: Output of the function (ranges from **0** to **1**)
- $x$: Input value (weighted sum of inputs in a neural network)
- $e$: Euler's number (approximately **2.71828**)

````python
def sig(u):
    result = []
    for i in u:
        E = math.e
        sigma = 1/(1 + (E**(-i)))
        result.append(sigma)
    return result
````

````python
g = sig(v)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Sigmoid Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sigmoid Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

 Binary Sigmoid Activation Function

It is mathematically defined as:

$$
\sigma(x) = \frac{1}{1 + e^{- \lambda x}}
$$

Where:

- $\sigma(x)$: Output of the function (ranges from **0** to **1**)
- $x$: Input value (weighted sum of inputs in a neural network)
- $e$: Euler's number (approximately **2.71828**)
- $\lambda$ : Any value in range 0 to 1

````python
def sig_bin(u, lam):
    result = []
    for i in u:
        E = math.e
        sigma = 1/(1 + (E**(-i * lam)))
        result.append(sigma)
    return result
````

````python
g = sig_bin(v, 0.3)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Binary Sigmoid Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Binary Sigmoid Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

 Bipolar Sigmoid Function

It is mathematically defined as:

$$
f(x)=\frac{1-e^{-\lambda x}}{1+e^{-\lambda x}}
$$

Where:

- $x$: Input value (the weighted sum of inputs plus bias)
- $e$: Euler's number (approximately **2.71828**)
- $\lambda$: Steepness (or slope) parameter of the sigmoid curve
- $f(x)$: Output of the function (ranges from **-1** to **1**)

````python
def sig_bpl(u, lam):
    result = []
    for i in u:
        E = math.e
        sigma = (1 - (E**(-i * lam)))/(1 + (E**(-i * lam)))
        result.append(sigma)
    return result
````

````python
g = sig_bpl(v, 0.3)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Bipolar Sigmoid Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bipolar Sigmoid Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

 Ramp Activation Function

It is mathematically defined as:

$$
f(x)=
\begin{cases}
0, & \text{for } x < 0 \\
x, & \text{for } 0 \le x \le 1 \\
1, & \text{for } x > 1
\end{cases}
$$

Where:

- $x$: Input value (the weighted sum of inputs plus bias)
- $f(x)$: Output of the function (ranges from **0** to **1**)
- For $x < 0$, the output is **0**
- For $0 \le x \le 1$, the output equals the input value ($x$)
- For $x > 1$, the output is **1**

````python
def ramp(u):
    result = []
    for i in u:
        if i < 0:
            result.append(0)
        elif i>= 0 and i <= 1:
            result.append(i)
        else:
            result.append(1)
    return result
````

````python
g = ramp(v)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Ramp Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ramp Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

 Tanh (Hyperbolic Tangent) Activation Function

It is mathematically defined as:

$$
f(x)=\tanh(x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}
$$

Where:

- $x$: Input value (the weighted sum of inputs plus bias)
- $f(x)$: Output of the function (ranges from **-1** to **1**)
- $e$: Euler's number (approximately **2.71828**)
- $\tanh(x)$: Hyperbolic tangent of the input value

````python
def tanh(u):
    result = []
    for i in u:
        E = math.e
        sigma = (E**i - (E**(-i)))/(E**i + (E**(-i)))
        result.append(sigma)
    return result
````

````python
g = tanh(v)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'TanH Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("TanH Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

ReLU (Rectified Linear Unit) Activation Function

The ReLU (Rectified Linear Unit) Activation Function is mathematically defined as:

$$
f(x)=\max(0,x)
$$

or equivalently,

$$
f(x)=
\begin{cases}
0, & \text{if } x < 0 \\
x, & \text{if } x \ge 0
\end{cases}
$$

Where:

- $x$: Input value (the weighted sum of inputs plus bias)
- $f(x)$: Output of the function (ranges from **0** to **∞**)
- For $x < 0$, the output is **0**
- For $x \ge 0$, the output equals the input value ($x$)

````python
def relu(u):
    results = []
    for i in u:
        results.append(max(0,i))
    return results
````

````python
g = relu(v)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Ramp Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ramp Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

Leaky ReLU (Leaky Rectified Linear Unit) Activation Function

It is mathematically defined as:

$$
f(x)=
\begin{cases}
x, & \text{if } x > 0 \\
\alpha x, & \text{if } x \leq 0
\end{cases}
$$

It can also be written as:

$$
f(x)=\max(\alpha x, x)
$$

Where:

- $x$: Input value (the weighted sum of inputs plus bias)
- $f(x)$: Output of the function
- $\alpha$: Small constant hyperparameter that controls the slope for negative inputs (typically **0.01**)
- For $x > 0$, the output is the input value ($x$)
- For $x \leq 0$, the output is a small negative value ($\alpha x$)

````python
def licky_relu(u, alpha = 0.01):
    results = []
    for i in u:
        results.append(max(alpha*i ,i))
    return results
````

````python
g = licky_relu(v)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Ramp Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ramp Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

Exponential Linear Unit (ELU) Activation Function

It is mathematically defined as:

$$
f(x)=
\begin{cases}
x, & \text{if } x > 0 \\
\alpha(\exp(x)-1), & \text{if } x \leq 0
\end{cases}
$$

Where:

- $x$: Input value (the weighted sum of inputs plus bias)
- $f(x)$: Output of the function
- $\alpha$: Positive constant that controls the value to which the function saturates for negative inputs (typically **1.0**)
- $\exp(x)$: Exponential function, equivalent to $e^x$
- For $x > 0$, the output is the input value ($x$)
- For $x \leq 0$, the output approaches $-\alpha$ as $x$ becomes more negative

````python
def elu(u, alpha):
    result = []
    E = math.e
    for i in u:
        if i > 0:
            result.append(i)
        else:
            result.append(alpha*((E**i) -1 ))
    return result
````

````python
g = elu(v, alpha= 0.1)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'ELU Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("ELU Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````

 Softmax Activation Function

It is mathematically defined as:

$$
\sigma(z_i)=\frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
$$

Where:

- $z_i$: Input score (logit) for the $i^{th}$ class
- $\sigma(z_i)$: Softmax output probability for the $i^{th}$ class
- $e$: Euler's number (approximately **2.71828**)
- $K$: Total number of classes
- $\sum_{j=1}^{K} e^{z_j}$: Sum of exponential values of all class scores

The output values are in the range **0 to 1**, and the sum of all output probabilities is equal to **1**.

````python
def soft(u):
    x = 0
    E = math.e
    result = []
    for i in u:
        x = x + E**i
    
    for i in u: 
        result.append((E**i)/x)
    return result
````

````python
g = soft(v)

plt.scatter(numbers,v, color = 'red', label = 'Random')
plt.scatter(numbers, g, color = 'blue', label = 'Soft Max Fucntion')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Soft Max Function Plot")
plt.legend()
plt.grid(True)

plt.show()
````
