# Detailed Implementation of Feed Forward ANN using List only

Importing Libraries

```python
import math
import random
```

Sigmoid activation function

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

Initialize Weights and Biases

```python
# Input (3) -> Hidden Layer 1 (4)
w1 = [[random.uniform(-1, 1) for j in range(4)] for i in range(3)]
b1 = [random.uniform(-1, 1) for i in range(4)]

# Hidden Layer 1 (4) -> Hidden Layer 2 (4)
w2 = [[random.uniform(-1, 1) for j in range(4)] for i in range(4)]
b2 = [random.uniform(-1, 1) for i in range(4)]

# Hidden Layer 2 (4) -> Output (1)
w3 = [[random.uniform(-1, 1)] for i in range(4)]
b3 = random.uniform(-1, 1)
```

```python
inputs = [0.5, 0.8, 0.2]
```

Hidden Layer 1

```python
hidden1 = []

for j in range(4):
    total = b1[j]
    for i in range(3):
        total += inputs[i] * w1[i][j]
    hidden1.append(sigmoid(total))
```

Hidden Layer 2

```python
hidden2 = []

for j in range(4):
    total = b2[j]
    for i in range(4):
        total += hidden1[i] * w2[i][j]
    hidden2.append(sigmoid(total))
```

Output Layer

```python
total = b3

for i in range(4):
    total += hidden2[i] * w3[i][0]

output = sigmoid(total)
```

```python
print("Input Layer :", inputs)
print("Hidden Layer 1 :", hidden1)
print("Hidden Layer 2 :", hidden2)
print("Output :", output)
```
