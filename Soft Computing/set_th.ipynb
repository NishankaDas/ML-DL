# Descriptive Implementation of Set Oporations 

This IPYNB will contain follwing set oporations :

* Union 
* Intersection 
* Diffrence
* Complement 
* Subset 
* Proper Subset

```python
universal_set = [5,1,7,2,9,3,70,5,36,47,99,10,29,38,45,16,80]
subset1 = universal_set[:7] + universal_set[12:]
subset2 = universal_set[7:]
```

Union 

```python
u = []
for i in subset1:
    if i not in u:
        u.append(i)
for i in subset2:
    if i not in u:
        u.append(i)

u.sort()
```

```python
u
```

Intersection

```python
inter = []
for i in subset1:
    if i in subset2 and i not in inter:
        inter.append(i)
```

```python
inter
```

Diffrence 

```python
diff = []
for i in subset1:
    if i not in subset2:
        diff.append(i)
```

```python
diff
```

Complement

```python
comp = []
for i in universal_set:
    if i not in subset1:
        comp.append(i)
```
