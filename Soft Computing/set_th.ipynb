# Descriptive Implementation of Set Oporations 

This IPYNB will contain follwing set oporations :

* Union 
* Intersection 
* Diffrence
* Complement 
* Subset 
* Proper Subset

```python
universal_set = [5,1,7,2,9,3,70,5,36,47,99,13,29,38,45,16,80]
subset1 = universal_set[:7] + universal_set[12:]
subset2 = universal_set[7:]
```

```python
import matplotlib.pyplot as plt
v = list(range(len(subset1)))
u = list(range(len(subset2)))
plt.scatter(subset2,u, color = "red", label="Subset2")
plt.scatter(subset1,v, color = "blue", label="Subset1")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

universal_set = [5,1,7,2,9,3,70,5,36,47,99,13,29,38,45,16,80]

subset1 = set(universal_set[:7] + universal_set[12:])
subset2 = set(universal_set[7:])

venn2([subset1, subset2], set_labels=('Group A', 'Group B'))

plt.title("2-Group Venn Diagram (From Sets)")
plt.show()
```

Union 

```python
u = []
for i in subset1:
    if i not in u:
        u.append(i)
for i in subset2:
    if i not in u:
        u.append(i)

u.sort()
```

```python
u
```

Intersection

```python
inter = []
for i in subset1:
    if i in subset2 and i not in inter:
        inter.append(i)
```

```python
inter
```

```python
x = list(range(len(inter)))
v = list(range(len(subset1)))
u = list(range(len(subset2)))
plt.scatter(subset2,u, color = "red", label="Subset2")
plt.scatter(subset1,v, color = "blue", label="Subset1")
plt.scatter(inter,x, color = "green", label="intersection")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot")
plt.legend()
plt.grid(True)
plt.show()
```

Diffrence 

```python
diff = []
for i in subset1:
    if i not in subset2:
        diff.append(i)
```

```python
diff
```

```python
x = list(range(len(diff)))
v = list(range(len(subset1)))
u = list(range(len(subset2)))
plt.scatter(subset2,u, color = "red", label="Subset2")
plt.scatter(subset1,v, color = "blue", label="Subset1")
plt.scatter(diff,x, color = "green", label="diffrence")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot")
plt.legend()
plt.grid(True)
plt.show()
```

Complement

```python
comp = []
for i in universal_set:
    if i not in subset1:
        comp.append(i)
```

```python
comp
```

```python
x = list(range(len(comp)))
v = list(range(len(subset1)))
plt.scatter(subset1,v, color = "blue", label="Subset1")
plt.scatter(comp,x, color = "green", label="diffrence")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot")
plt.legend()
plt.grid(True)
plt.show()
```
