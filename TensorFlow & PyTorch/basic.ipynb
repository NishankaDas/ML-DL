```python
import numpy as np
def fibo(n):
    k = 1/ np.sqrt(5)
    j = ((1 + np.sqrt(5))/2)**n
    l = ((1 - np.sqrt(5))/2)**n
    return int(k*(j-l))
```

```python
print(fibo(55))
```
