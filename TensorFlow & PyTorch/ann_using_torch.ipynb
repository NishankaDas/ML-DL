# Descriptive Implementation of ANN using PyTorch

Importing Libraries

```python
import torch
```

Sigmoid Function

```python
sigmoid = torch.sigmoid
```

Initialize Weights and Biases

```python
# Input (3) -> Hidden Layer 1 (4)
W1 = torch.rand(3, 4)
b1 = torch.rand(1, 4)

# Hidden Layer 1 (4) -> Hidden Layer 2 (4)
W2 = torch.rand(4, 4)
b2 = torch.rand(1, 4)

# Hidden Layer 2 (4) -> Output (1)
W3 = torch.rand(4, 1)
b3 = torch.rand(1, 1)
```

Initial Value 

```python
X = torch.tensor([[0.5, 0.8, 0.2]])
```

Feed Forward

```python
# Hidden Layer 1
H1 = sigmoid(torch.matmul(X, W1) + b1)

# Hidden Layer 2
H2 = sigmoid(torch.matmul(H1, W2) + b2)

# Output Layer
Y = sigmoid(torch.matmul(H2, W3) + b3)
```

Display Result

```python
print("Input:")
print(X)

print("\nHidden Layer 1:")
print(H1)

print("\nHidden Layer 2:")
print(H2)

print("\nOutput:")
print(Y)
```
