# Descriptive Implentation of Fit Forward ANN

The basic structure of the ANN includes requires multiple muliplication of input layers with the given weights, here the bias is now used in this implementation. And for the last layer the addative sum of is given as output, after that we can use sigmoid function to get the value in binary.

Importing required packges for fit forward in ANN

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Enter the number of Inputs and Layers

```python
n = 3 # Inputs
l = 4 # Layers
```

Input the Input vector here

```python
input_l = [0.3,0.7,0.9] #Input vector
lists = [[0] * n for _ in range(l)]
```

```python
lists
```

Weight vectors for the layers

```python
for j in range(0,l):
    for i in range(0,n):
        lists[j][i] = round(random.random(), 3)
```

```python
lists
```

Values of nodes in every layers

```python
print(input_l)
print("ANN Starts Here")
for i in range(l):
    weight = np.array(lists[i])
    value = np.array(input_l)
    input_l = weight * value
    print(input_l)
```

Last layer where the nodes marge into one

```python
last_layer = np.array(input_l)

weights = np.random.rand(len(last_layer))

output = np.dot(weights, last_layer)
```
